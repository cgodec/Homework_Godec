```python
import pandas as pd
df = pd.read_csv(r'C:\Users\crgod\anaconda_projects\fwe\data\MaunaLoaCO2_monthly_sitename.csv') #Imported data using file path
df.head() #Display first 5 rows
```

```python
df1 = df[(df.Year !=1958) & #Set df1 equal to df where Year is not 1958
(df.Year!=2022) & #Set df1 equal to df where Year is not 2022
(df.numofdays>=0) & #Set df1 equal to df where Number of days is not negative
(df.stdofdays>=0)] #Set df1 equal to df where the std of days is not negative
df1
```

```python
df1['annual_average'] = df1.groupby('Year').Monthly_average.transform('mean') #Group by year and calculate mean of monthly average
df1['deseasonalized_annual_average'] = df.groupby('Year').deseasonalized.transform('mean') #Group by year and calculate mean of deseasonalized
df1['std_of_months'] = df.groupby('Year').Monthly_average.transform('std') #Group by year and calculate standard deviation of monthly average
df1
```

```python
def CO2(val): #Define function
    if val < 350: 
        return 'Low' #If value is less than 350 return 'Low'
    elif val >= 350 and val <= 400: 
        return 'Moderate' #If value is greater than or equal to 350 and less than or equal to 400 return 'Moderate'
    else:
        return 'High' #If else return 'High'

df1['CO2_Category'] = df1.Monthly_average.apply(CO2) #Apply C02 function and store in new column 'CO2_Category' 
df1
    
```

```python
df1.to_csv('Godec_MaunaLoaCO2_annual_Hw3.csv', index=False)
```
